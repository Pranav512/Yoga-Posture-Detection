```python
from tensorflow.keras.models import load_model
```

```python
import tensorflow as tf
import os
import cv2
import imghdr
import numpy as np
from matplotlib import pyplot as plt
```

```python
new_model = load_model(os.path.join('models','trainedmodel1.h5'))
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()

    # Recolor Feed
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    image = cv2.flip(image, 1)
    image.flags.writeable = False
    
    results = tf.image.resize(image, (256,256))
    y_res = new_model.predict(np.expand_dims(results/255, 0))
    y_class = [np.argmax(element) for element in y_res]
    
    if y_res[0][y_class]<0.8:
        txt='Try Again'
    else:
        if int(y_class[0])==0:
            txt='Mountain'
        elif int(y_class[0])==1:
            txt='Tree'
        elif int(y_class[0])==2:
            txt='Lotus'
        elif int(y_class[0])==3:
            txt='Lunge'
        elif int(y_class[0])==4:
            txt='Plank'
        elif int(y_class[0])==5:
            txt='DownDog'
    
    # Recolor image back to BGR for rendering
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    image.flags.writeable = True
    
    # Get status box
    cv2.rectangle(image, (0,0), (640, 40), (225, 0, 144), -1)
    
    # Display Class
    cv2.putText(image, 'Class : ', (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
    cv2.putText(image, txt, (80, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1, cv2.LINE_AA)
    
    #Display Probability
    cv2.putText(image, 'Prob : '
               , (160,25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
    cv2.putText(image, str(y_res[0][y_class])
               , (220,25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1, cv2.LINE_AA)

    cv2.imshow('Raw Webcam Feed', image)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
